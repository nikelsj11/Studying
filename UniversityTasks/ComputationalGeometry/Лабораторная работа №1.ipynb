```python
from ipywidgets import interact
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import axes
from matplotlib import rcParams, rc
```

## *Задание №1*

```python
rcParams['figure.figsize'] = (10.0, 5.0)
rcParams['figure.dpi'] = 190
rcParams['font.size'] = 14

function_to_plot = lambda x: -x**2 + x + 1
x = np.arange(-4.,4.,0.01)

diff_point = dict(x = -3, y = function_to_plot(-3))
max_point = dict( x = None, y = 0)

for i in x:
    rec = function_to_plot(i)
    if rec > max_point['y']:
        max_point['x'] = i
        max_point['y'] = rec

plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.title(r'Plot $y(x)=-x**2+x+1, x=4.00,...,-4.00$')
plt.grid(True)

ax = plt.subplot(111)
ax.set_xlim(-4., 4.)
ax.set_ylim(-20., 5.)

ax.annotate(
    'Maximum point', 
    (max_point['x'], max_point['y']), 
    xytext=(max_point['x']-1.5, max_point['y']+1.5),
    arrowprops=dict(facecolor='black',  arrowstyle="->")
)

ax.annotate(
    r'$dy(x_0)/dx=-11.00$', 
    (diff_point['x'], diff_point['y']), 
    xytext=(diff_point['x']+2, diff_point['y']+0.3),    
    bbox=dict(boxstyle="round", fc="0.8"),
    arrowprops=dict(facecolor='black',  arrowstyle="->",),
)

plt.text(-1.5, -25, 'Fig. 1: Serial bubble sort algorithm',)

ax.plot(x,-x**2 + x + 1)

points_x = [max_point['x'], diff_point['x']]
points_y = [max_point['y'], diff_point['y']]
ax.plot(points_x, points_y, 'ko')


```

## *Задание №3*

```python
rcParams['figure.figsize'] = (20.0, 10.0)
rcParams['figure.dpi'] = 190
rcParams['font.size'] = 12

def re_plot(a,b):
    fig = plt.figure(0)
    ax = fig.add_subplot(1,1,1)
    ax.set_aspect
    ax.set_xlim(-10.0,10.0)
    ax.set_ylim(-10.0,10.0)
    t1 = np.arange(0,2*np.pi,0.01)
    t2 = np.arange(-10.0,10.0,0.01)
    ax.set_title("a ={0:5.2f},b ={1:5.2f}".format(a,b))
    ax.plot(a*np.cos(t1),b*np.sin(t1))
    ax.plot(a*np.cosh(t2),b*np.sinh(t2))
    
_ = interact(re_plot, a = (0.0,10.0,0.1), b = (0.0,10.0,0.1))
```

## *Задание №4*

```python
import csv

# наша функция от одного параметра и одного аргумента 
some_function = lambda x, t: x**t

# пишем CVS файл
with open('dist/some_function.csv', 'w') as some_function_file:
    csv_writer = csv.writer(some_function_file)
    for x in range(0, 5):
        for t in range(0, 5):
            csv_writer.writerow([x, t, some_function(x,t)])

# читаем CVS файл
x, t, y = [], [], []

with open('dist/some_function.csv', 'r') as some_function_file:
    csv_reader = csv.reader(some_function_file)
    for row in csv_reader:
        x.append(row[0])
        t.append(row[1])
        y.append(row[2])
        
print("===> Прочитанный CVS файл ~native~: ")
print("Массив X: ", np.array(x))
print("Массив T: ", np.array(t))
print("Массив Y: ", np.array(y))

print("\n\n")

# читаем CVS файл используя NumPy
some_function_file = np.genfromtxt ('dist/some_function.csv', delimiter=",")

print("===> Прочитанный c CVS файл ~numpy~: ")
print("Массив X: ", some_function_file[:,0])
print("Массив T: ", some_function_file[:,1])
print("Массив Y: ", some_function_file[:,2])
print("Тип массивов: ", type(some_function_file[:,0]))
```
